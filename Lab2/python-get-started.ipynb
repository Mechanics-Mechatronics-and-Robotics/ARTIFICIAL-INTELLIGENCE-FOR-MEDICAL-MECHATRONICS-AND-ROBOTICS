# Кортежи

```python
# Кортежи в функциях

def find_max(x):
    ind = 0
    maximum = x[0]
    for i in range(1,len(x)):
        if x[i] > maximum:
            maximum = x[i]
            ind = i
    return maximum, ind

out = find_max([4,3,5])

print(out)
print(type(out))
print(out[1])
```

```python
# Распаковка кортежа

maximum, index = find_max([4,3,5])

print(maximum, index)
```

# Линейная регрессия с помощью автоматического дифференцирования

```python
from matplotlib import pyplot as plt

x = torch.tensor([24, 27, 32, 37, 41, 43, 50, 52], dtype=torch.float32)
y = torch.tensor([17, 18.8, 18, 19.5, 21, 20, 27, 29], dtype=torch.float32)
a = torch.randn((1), dtype = torch.float32, requires_grad=True)
b = torch.randn((1), dtype = torch.float32, requires_grad=True)
lr = 0.00005

y_plot = a * x + b

fig, ax = plt.subplots(figsize=(8, 5), ncols=1)
ax.scatter(x, y)
ax.plot(x.detach(), y_plot.detach(), c='r')

plt.show()
```

```python
history = []

for i in range(30):
    if a.grad:
        a.grad.data.zero_()
        b.grad.data.zero_()
    
    out = a * x + b
    
    loss = (y - out).square().mean()
    loss.backward()
    
    history.append(loss.item())
    
    print('Loss:', loss)
    
    with torch.no_grad():
        a.sub_(a.grad.data * lr)
        b.sub_(b.grad.data * lr)
```

```python
y_plot = a * x + b

fig, ax = plt.subplots(figsize = (16, 5), ncols=2)
ax[0].scatter(x, y)
ax[0].plot(x.detach(), y_plot.detach(), c='r')
ax[1].plot(history, c='g')

plt.show()
```

# Линейная регрессия при помощи средств Pytorch

```python
class LinRegDataset(torch.utils.data.Dataset):
    
    def __init__(self, x, y):
        self.x = x.unsqueeze(1)
        self.y = y.unsqueeze(1)

    def __len__(self):
        return len(self.x)

    def __getitem__(self, idx):
        return (self.x[idx], self.y[idx])

class LenearRegression(torch.nn.Module):

    def __init__(self, number_of_features, number_of_outputs, dropout_rate=0):
        super(LenearRegression, self).__init__()
        self.layers = torch.nn.Sequential(torch.nn.Dropout(dropout_rate),
                                          torch.nn.Linear(number_of_features, number_of_outputs, bias=True))

    def forward(self, x):
        return self.layers(x)

x = torch.tensor([24, 27, 32, 37, 41, 43, 50, 52], dtype=torch.float32)
y = torch.tensor([17, 18.8, 18, 19.5, 21, 20, 27, 29], dtype=torch.float32)

dataset = LinRegDataset(x, y)

loader = torch.utils.data.DataLoader(dataset, batch_size=2)

criterion = torch.nn.MSELoss()

model = LenearRegression(1, 1)

optimizer = torch.optim.SGD(model.parameters(), lr=0.00005)

y_plot = model.forward(dataset.x)

fig, ax = plt.subplots(figsize=(8, 5), ncols=1)
ax.scatter(x, y)
ax.plot(x.detach(), y_plot.detach(), c='r')

plt.show()
```

```python
history = []

if torch.cuda.is_available():
    model = model.to('cuda')

for i in range(30):
    run_loss = 0
    iters = 0
    
    for x_batch, y_batch in loader:
        if torch.cuda.is_available():
            x_batch = x_batch.to('cuda')
            y_batch = y_batch.to('cuda')
        
        optimizer.zero_grad()
        
        out = model.forward(x_batch)
        
        loss = criterion(out, y_batch)
        loss.backward()
        
        optimizer.step()
        
        iters += 1
        
        run_loss += loss.item()
    
    history.append(run_loss / iters)
    
    print('Loss:', run_loss / iters)
```

```python
y_plot = model.forward(dataset.x)

fig, ax = plt.subplots(figsize=(16, 5), ncols=2)
ax[0].scatter(x, y)
ax[0].plot(x.detach(), y_plot.detach(), c='r')
ax[1].plot(history, c='g')

plt.show()
```
